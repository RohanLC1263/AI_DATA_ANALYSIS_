```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load iris dataset
iris = sns.load_dataset('iris')

# Select only numeric columns for correlation
numeric_cols = iris.select_dtypes(include=['float64', 'int64'])

# Compute correlation matrix on numeric columns only
corr = numeric_cols.corr()

# Create heatmap with advanced styling
plt.figure(figsize=(8, 6))
sns.heatmap(
    corr,
    annot=True,
    fmt=".2f",
    cmap='coolwarm',
    linewidths=0.5,
    linecolor='white',
    cbar_kws={"shrink": 0.8, "label": "Correlation Coefficient"},
    square=True,
    annot_kws={"size": 12, "weight": 'bold', "color": 'black'}
)

plt.title('Correlation Matrix Heatmap of Iris Dataset (Numeric Features)', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```
