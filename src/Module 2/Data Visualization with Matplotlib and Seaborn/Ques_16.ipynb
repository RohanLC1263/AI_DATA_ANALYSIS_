```python
# Combining Plots with Matplotlib
# Objective: Combine multiple plots into a single figure using Matplotlib's subplots.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.arange(1, 6)
y_line = [10, 15, 13, 18, 20]
y_bar = [5, 7, 9, 6, 8]

# Create subplots (1 row, 2 columns)
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Line plot on the first subplot
axs[0].plot(x, y_line, marker='o', linestyle='-', color='blue')
axs[0].set_title('Line Plot')
axs[0].set_xlabel('X-axis')
axs[0].set_ylabel('Y-axis')

# Bar plot on the second subplot
axs[1].bar(x, y_bar, color='orange')
axs[1].set_title('Bar Plot')
axs[1].set_xlabel('X-axis')
axs[1].set_ylabel('Values')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show combined plot
plt.show()
```
