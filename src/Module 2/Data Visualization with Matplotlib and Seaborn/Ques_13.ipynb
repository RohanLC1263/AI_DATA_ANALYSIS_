```python
# Customizing Plots with Seaborn - Boxplot
# Objective: Customize a boxplot with Seaborn to show the distribution of petal lengths by
# species.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load iris dataset
iris = sns.load_dataset('iris')

# Create customized boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x='species', y='petal_length', data=iris, palette='Set2')

# Customize title and labels
plt.title('Distribution of Petal Length by Species', fontsize=16, fontweight='bold')
plt.xlabel('Species', fontsize=14)
plt.ylabel('Petal Length (cm)', fontsize=14)

plt.tight_layout()
plt.show()
```
