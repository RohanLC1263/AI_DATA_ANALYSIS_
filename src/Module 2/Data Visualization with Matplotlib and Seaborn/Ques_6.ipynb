```python
# Customizing Matplotlib Plot Styles
# Description: Customize the Matplotlib bar plot from Question 1 with a grid, different color, and custom fonts.
```

```python
import matplotlib.pyplot as plt

# Data
products = ['A', 'B', 'C', 'D', 'E']
sales = [23, 75, 57, 90, 65]

# Customize fonts
title_font = {'family': 'serif', 'color': 'darkred', 'weight': 'bold', 'size': 16}
label_font = {'family': 'serif', 'color': 'darkblue', 'weight': 'normal', 'size': 12}

# Plot
plt.figure(figsize=(8, 5))
plt.bar(products, sales, color='teal', edgecolor='black')

# Add grid
plt.grid(axis='y', linestyle='--', linewidth=0.7, alpha=0.7)

# Add labels and title with custom fonts
plt.title('Product Sales Bar Plot (Customized)', fontdict=title_font)
plt.xlabel('Products', fontdict=label_font)
plt.ylabel('Sales (in thousands)', fontdict=label_font)

plt.tight_layout()
plt.show()
```
