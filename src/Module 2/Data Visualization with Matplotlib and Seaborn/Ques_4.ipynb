```python
# Basic Seaborn Bar Plot
# Description: Using Seaborn, plot the same bar plot from Question 1 to compare the sales data.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Step 1: Prepare the data in a DataFrame
data = {
    'Product': ['A', 'B', 'C', 'D', 'E'],
    'Sales': [23, 75, 57, 90, 65]  # Sales in thousands
}
df = pd.DataFrame(data)

# Step 2: Create the bar plot using Seaborn
plt.figure(figsize=(8, 5))
sns.barplot(x='Product', y='Sales', data=df, palette='pastel')

# Step 3: Add title and labels
plt.title('Product Sales Bar Plot (Seaborn)')
plt.xlabel('Products')
plt.ylabel('Sales (in thousands)')

# Step 4: Show the plot
plt.tight_layout()
plt.show()
```
