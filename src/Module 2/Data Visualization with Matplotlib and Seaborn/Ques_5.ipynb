```python
# Basic Seaborn Scatter Plot
# Description: Using Seaborn, recreate the scatter plot from Question 2 for height vs. weight.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Step 1: Prepare the data in a DataFrame
data = {
    'Height_cm': [160, 170, 165, 180, 175],
    'Weight_kg': [55, 65, 60, 80, 76]
}
df = pd.DataFrame(data)

# Step 2: Create the scatter plot using Seaborn
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Height_cm', y='Weight_kg', data=df, color='blue', s=100)

# Step 3: Add title and labels
plt.title('Height vs Weight Scatter Plot (Seaborn)')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')

# Step 4: Show the plot
plt.tight_layout()
plt.show()
```
