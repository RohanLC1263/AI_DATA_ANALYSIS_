```python
# Styling with Matplotlib - Line Plot
# Objective: Create a line plot with customized styles using Matplotlib.
```

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 18, 20]

# Customize fonts for title and labels
title_font = {'fontsize': 16, 'fontweight': 'bold', 'color': 'navy'}
label_font = {'fontsize': 12, 'color': 'darkgreen'}

# Plot with custom style
plt.figure(figsize=(8, 5))
plt.plot(x, y, color='purple', linestyle='--', marker='o', markerfacecolor='orange', markersize=8, linewidth=2)

# Add title and labels with custom fonts
plt.title('Custom Styled Line Plot', fontdict=title_font)
plt.xlabel('X-axis Label', fontdict=label_font)
plt.ylabel('Y-axis Label', fontdict=label_font)

# Add grid with custom style
plt.grid(True, linestyle=':', linewidth=1, color='gray', alpha=0.7)

plt.tight_layout()
plt.show()
```
