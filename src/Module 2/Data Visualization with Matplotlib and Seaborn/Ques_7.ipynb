```python
# Advanced Scatter Plot with Seaborn
# Description: Using Seaborn, enhance the scatter plot for height vs. weight by adding size for points based
# on data: Size [10, 15, 13, 18, 12].
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Data with Size for points
data = {
    'Height_cm': [160, 170, 165, 180, 175],
    'Weight_kg': [55, 65, 60, 80, 76],
    'Size': [10, 15, 13, 18, 12]
}
df = pd.DataFrame(data)

# Plot scatter with varying sizes
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Height_cm', y='Weight_kg', size='Size', sizes=(50, 300), data=df, legend='brief', palette='viridis')

# Add labels and title
plt.title('Height vs Weight Scatter Plot with Point Sizes')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')

plt.tight_layout()
plt.show()
```
