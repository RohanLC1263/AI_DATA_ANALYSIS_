```python
# Advanced Seaborn Plot - Pairplot
# Objective: Use Seaborn to create a pairplot to visualize pairwise relationships in a dataset.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris = sns.load_dataset('iris')

# Create pairplot
sns.pairplot(iris, hue='species', palette='bright', diag_kind='kde')

# Show plot
plt.suptitle('Pairwise Relationships in Iris Dataset', y=1.02)  # Title above the plot
plt.show()
```
