```python
# Title: Feature Engineering for Classification
# Description: Create a new feature that could help distinguish between species based on
# logical assumptions and verify its utility.
```

```python
import pandas as pd
from sklearn.datasets import load_iris
import seaborn as sns
import matplotlib.pyplot as plt

# Load Iris dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# Logical assumption:
# Petal length and petal width tend to be larger for versicolor and virginica species.
# Create a new feature: petal_area = petal length * petal width
df['petal_area'] = df['petal length (cm)'] * df['petal width (cm)']

# Check if 'petal_area' helps distinguish species by visualizing distributions
plt.figure(figsize=(8, 6))
sns.boxplot(x='species', y='petal_area', data=df, palette='Set2')
plt.title('Petal Area Distribution by Species')
plt.ylabel('Petal Area (cm²)')
plt.show()

# Summary statistics of the new feature by species
print("Summary statistics of petal_area by species:")
print(df.groupby('species')['petal_area'].describe())
```
