```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset from seaborn's built-in datasets
iris = sns.load_dataset("iris")

# Set seaborn style for better aesthetics
sns.set(style="ticks", color_codes=True)

# Create a pair plot showing pairwise relationships colored by species
pair_plot = sns.pairplot(iris, hue="species", diag_kind="kde", palette="husl")

# Add a title to the plot
pair_plot.fig.suptitle("Pair Plot of Iris Dataset Features by Species", y=1.02)

# Display the plot
plt.show()
```
