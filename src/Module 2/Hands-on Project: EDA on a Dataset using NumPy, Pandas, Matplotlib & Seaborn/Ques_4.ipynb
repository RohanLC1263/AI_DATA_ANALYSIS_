```python
# Title: Basic Data Visualization
# Description: Plot the distribution of one feature using Seaborn.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset from seaborn's built-in datasets
df = sns.load_dataset("iris")

# Plot the distribution of the 'sepal_length' feature using a histogram
sns.histplot(data=df, x="sepal_length", kde=True)

plt.title("Distribution of Sepal Length in Iris Dataset")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")

plt.show()
```
