```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset from seaborn's built-in datasets
iris = sns.load_dataset("iris")

# Create a pair plot to display pairwise relationships
sns.pairplot(data=iris, hue="species", diag_kind="kde", palette="husl")

plt.suptitle("Iris Dataset Pair Plot", y=1.02)  # Adjust title position

plt.show()
```
