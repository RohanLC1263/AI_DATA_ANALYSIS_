```python
import pandas as pd
import numpy as np

# ----------------------------
# Part 1: Handle Missing Values & Duplicates
# ----------------------------

# 1. Load Data (using a sample dataset here)
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Alice', None, 'Eve', 'Frank', 'Bob'],
    'Age': [25, np.nan, 30, 25, 40, np.nan, 35, np.nan],
    'Email': ['alice@example.com', 'bob@example.com', 'charlie@example.com', 'alice@example.com', None, 'eve@example.com', 'frank@example.com', 'bob@example.com']
}

df = pd.DataFrame(data)
print("Original Data:")
print(df)

# 2. Handling Missing Values

# 2.1 Identify Missing Values
print("\nMissing values per column:")
print(df.isnull().sum())

# 2.2 Fill Missing Values
# Fill missing 'Age' with median age
median_age = df['Age'].median()
df['Age'] = df['Age'].fillna(median_age)

# Fill missing 'Name' with placeholder 'Unknown'
df['Name'] = df['Name'].fillna('Unknown')

# 3. Handling Duplicates

# 3.1 Identify Duplicates
duplicates = df.duplicated()
print("\nDuplicate rows:")
print(df[duplicates])

# 3.2 Remove Duplicates
df = df.drop_duplicates()
print("\nData after removing duplicates:")
print(df)

# ----------------------------
# Part 2: Apply Standardization & Formatting Rules
# ----------------------------

# 1. Standardize Text Data

# 1.1 Convert all 'Name' values to lowercase
df['Name'] = df['Name'].str.lower()

# 2. Format Numerical Data

# 2.1 Round 'Age' to nearest integer
df['Age'] = df['Age'].round().astype(int)

print("\nData after standardizing text and formatting numbers:")
print(df)

# 3. Combined Practice on Another Dataset

# 3.1 New Sample Dataset for products
product_data = {
    'Product': ['Laptop', 'SmartPhone', 'HeadPhones', 'Laptop', 'smartphone'],
    'Price': [999.999, 499.499, 79.995, 999.999, 499.499]
}

product_df = pd.DataFrame(product_data)
print("\nOriginal Product Data:")
print(product_df)

# 3.2 Standardize Product Names (lowercase)
product_df['Product'] = product_df['Product'].str.lower()

# 3.3 Format Prices to 2 decimal places
product_df['Price'] = product_df['Price'].round(2)

print("\nProduct Data after standardization and formatting:")
print(product_df)
```
