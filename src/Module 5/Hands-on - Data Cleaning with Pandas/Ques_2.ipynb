```python
import pandas as pd

# ---------------------------------------------
# Task 1: Remove Duplicates
# ---------------------------------------------

# 1. Load Extended Data (with intentional duplicates)
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Alice', 'Eve', 'Bob'],
    'Age': ['25', '30', '35', '25', '28', '30'],
    'JoinDate': ['2023-01-15', '2023-02-20', '2023-03-10', '2023-01-15', '2023-04-05', '2023-02-20']
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

# 2. Remove Duplicates
df_no_duplicates = df.drop_duplicates()

print("\nDataFrame After Removing Duplicates:")
print(df_no_duplicates)

# ---------------------------------------------
# Task 2: Fix Data Types
# ---------------------------------------------

# 1. Fix Incorrect Data Types (e.g., 'Age' should be numeric)
df_no_duplicates['Age'] = pd.to_numeric(df_no_duplicates['Age'], errors='coerce')

print("\nDataFrame After Fixing Data Types (Age to Numeric):")
print(df_no_duplicates.dtypes)
print(df_no_duplicates)

# ---------------------------------------------
# Task 3: Convert Data Type for Analysis
# ---------------------------------------------

# 1. Convert Date Strings to DateTime
df_no_duplicates['JoinDate'] = pd.to_datetime(df_no_duplicates['JoinDate'], format='%Y-%m-%d', errors='coerce')

print("\nDataFrame After Converting JoinDate to datetime:")
print(df_no_duplicates.dtypes)
print(df_no_duplicates)
```
