```python
import pandas as pd
import numpy as np

# -----------------------------
# Task 1: Detect Missing Data
# -----------------------------

# 1. Load the Data
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, np.nan, 35, 45, np.nan],
    'Gender': ['Female', 'Male', np.nan, 'Male', 'Female'],
    'Salary': [50000, 60000, np.nan, 80000, 75000]
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

# 2. Detect Missing Data
print("\nMissing Values (True if Missing):")
print(df.isnull())

print("\nCount of Missing Values Per Column:")
print(df.isnull().sum())


# --------------------------------------------
# Task 2: Handle Missing Data by Dropping
# --------------------------------------------

# 1. Drop Rows with Missing Values
df_dropped = df.dropna()

print("\nDataFrame After Dropping Rows with Missing Values:")
print(df_dropped)


# --------------------------------------------
# Task 3: Handle Missing Data by Imputation
# --------------------------------------------

# 1. Fill Missing Values
df_imputed = df.copy()

# Impute numerical columns with mean
df_imputed['Age'] = df_imputed['Age'].fillna(df_imputed['Age'].mean())
df_imputed['Salary'] = df_imputed['Salary'].fillna(df_imputed['Salary'].mean())

# Impute categorical column with mode
df_imputed['Gender'] = df_imputed['Gender'].fillna(df_imputed['Gender'].mode()[0])

print("\nDataFrame After Imputation:")
print(df_imputed)
```
