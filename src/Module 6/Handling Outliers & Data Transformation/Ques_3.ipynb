```python
# Question: Power Transformation (Box-Cox) Method
# Description: Use Box-Cox transformation to stabilize variance and make the data more normal.


```

```python
import pandas as pd
import numpy as np
from scipy.stats import boxcox
import seaborn as sns
import matplotlib.pyplot as plt

# Sample skewed data (e.g., exponential distribution)
np.random.seed(0)
data = np.random.exponential(scale=2.0, size=1000)
df = pd.DataFrame({'Skewed_Data': data})

# Visualize original data
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(df['Skewed_Data'], kde=True, color='red')
plt.title('Original Skewed Data')

# Apply Box-Cox transformation
# Add a small constant if minimum is 0 (Box-Cox requires all positive values)
boxcox_transformed, fitted_lambda = boxcox(df['Skewed_Data'] + 0.0001)

df['BoxCox_Transformed'] = boxcox_transformed

# Visualize transformed data
plt.subplot(1, 2, 2)
sns.histplot(df['BoxCox_Transformed'], kde=True, color='green')
plt.title(f'Box-Cox Transformed Data ( = {fitted_lambda:.2f})')

plt.tight_layout()
plt.show()
```
