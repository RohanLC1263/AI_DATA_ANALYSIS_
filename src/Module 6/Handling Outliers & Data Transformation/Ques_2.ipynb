```python
# Question: Winsorization of Data
# Description: Apply Winsorization to a given dataset to handle outliers.

```

```python
import pandas as pd
from scipy.stats.mstats import winsorize
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data with outliers
data = {
    'Income': [35000, 36000, 37000, 38000, 39000, 40000, 41000, 1000000, 42000, 43000, 44000, 45000, 500000]
}
df = pd.DataFrame(data)

# Apply Winsorization (e.g., limit 5% from both ends)
winsorized_data = winsorize(df['Income'], limits=[0.05, 0.05])

# Create a new DataFrame to compare
df['Winsorized_Income'] = winsorized_data

# Plot before and after
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.boxplot(x=df['Income'], color='orange')
plt.title("Original Data (with Outliers)")

plt.subplot(1, 2, 2)
sns.boxplot(x=df['Winsorized_Income'], color='green')
plt.title("Winsorized Data")

plt.tight_layout()
plt.show()

```
