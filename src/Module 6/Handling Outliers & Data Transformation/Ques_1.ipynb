```python
# Question: Visualizing Outliers Using Boxplots
# Description: Use a boxplot to visually detect outliers in a dataset.


```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataset with outliers
data = {
    'Scores': [75, 80, 82, 79, 81, 85, 83, 120, 78, 76, 77, 130, 74, 73, 72]
}
df = pd.DataFrame(data)

# Create a boxplot to visualize outliers
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['Scores'], color='skyblue')
plt.title('Boxplot of Scores')
plt.xlabel('Scores')
plt.grid(True)
plt.show()
```
