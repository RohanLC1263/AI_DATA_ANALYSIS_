```python
# Activity 2: Dealing with Duplicates & Redundancy

# Task A: Identifying Duplicate Records

# 7. Identify Complete Duplicates:
# - Load a dataset and identify duplicated rows.
# - Use Pandas to detect duplicates.






# 8. Identify Duplicates based on Specific Columns:
# - Check for duplicates in specified columns.







# 9. Count Duplicate Rows:
# - Calculate and print the number of duplicate rows.





```

```python
import pandas as pd

# Simulated dataset (you can replace this with: pd.read_csv("your_dataset.csv"))
data = {
    'EmployeeID': [101, 102, 103, 104, 101],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Alice'],
    'Department': ['HR', 'Finance', 'IT', 'HR', 'HR'],
    'Salary': [50000, 60000, 70000, 80000, 50000]
}
df = pd.DataFrame(data)
print("Original Dataset:")
print(df)

# ------------------------------------
# 7. Identify Complete Duplicates
# ------------------------------------
complete_duplicates = df[df.duplicated(keep=False)]
print("\nComplete Duplicate Rows (all columns match):")
print(complete_duplicates)

# ------------------------------------
# 8. Identify Duplicates based on Specific Columns
# ------------------------------------
duplicate_names = df[df.duplicated(subset=['Name', 'Salary'], keep=False)]
print("\nDuplicates Based on ['Name', 'Salary']:")
print(duplicate_names)

# ------------------------------------
# 9. Count Duplicate Rows
# ------------------------------------
duplicate_count = df.duplicated().sum()
print(f"\nNumber of Complete Duplicate Rows: {duplicate_count}")
```

```python
# Task B: Deduplication Techniques

# 10. Remove Complete Duplicates:
# - Drop duplicate rows and keep only the first occurrence.






# 11. Subset Deduplication:
# - Remove duplicates based on a subset of columns.






# 12. Keep Last Occurrence:
# - Drop duplicates but keep the last occurrence in the dataset.





```

```python
import pandas as pd

# Sample dataset (you can replace with pd.read_csv('employees.csv'))
data = {
    'EmployeeID': [101, 102, 103, 104, 101, 105],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Alice', 'Eve'],
    'Department': ['HR', 'Finance', 'IT', 'HR', 'HR', 'IT'],
    'Salary': [50000, 60000, 70000, 80000, 50000, 70000]
}
df = pd.DataFrame(data)
print("Original Dataset:")
print(df)

# -------------------------------------
# 10. Remove Complete Duplicates
# -------------------------------------
df_no_duplicates = df.drop_duplicates()
print("\nAfter Removing Complete Duplicates (Keep First):")
print(df_no_duplicates)

# -------------------------------------
# 11. Subset Deduplication (e.g., based on 'Name' and 'Salary')
# -------------------------------------
df_subset_dedup = df.drop_duplicates(subset=['Name', 'Salary'])
print("\nAfter Subset Deduplication (['Name', 'Salary']):")
print(df_subset_dedup)

# -------------------------------------
# 12. Keep Last Occurrence of Duplicates
# -------------------------------------
df_keep_last = df.drop_duplicates(keep='last')
print("\nAfter Removing Duplicates (Keep Last):")
print(df_keep_last)
```
