```python
# Question: Evaluating Data Distribution
# Description: Analyze the distribution of a numeric column using histograms and descriptive statistics.

```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Example: Load sample data (using Iris dataset's sepal length as numeric column)
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Choose a numeric column to analyze
column = 'sepal length (cm)'
data = df[column]

# Descriptive statistics
mean_val = np.mean(data)
median_val = np.median(data)
mode_val = data.mode()[0] if not data.mode().empty else None
variance_val = np.var(data)
std_dev_val = np.std(data)
min_val = np.min(data)
max_val = np.max(data)
range_val = max_val - min_val

print(f"Descriptive statistics for '{column}':")
print(f"Mean: {mean_val:.3f}")
print(f"Median: {median_val:.3f}")
print(f"Mode: {mode_val:.3f}")
print(f"Variance: {variance_val:.3f}")
print(f"Standard Deviation: {std_dev_val:.3f}")
print(f"Min: {min_val:.3f}")
print(f"Max: {max_val:.3f}")
print(f"Range: {range_val:.3f}")

# Plot histogram with seaborn
plt.figure(figsize=(8, 5))
sns.histplot(data, bins=15, kde=True)
plt.title(f"Distribution of {column}")
plt.xlabel(column)
plt.ylabel("Frequency")
plt.show()
```
