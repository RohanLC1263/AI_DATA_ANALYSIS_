```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset
df = sns.load_dataset('titanic')
print("First 5 rows of the Titanic dataset:")
print(df.head())

# Check for missing values in the 'age' column
print("\nMissing values in the 'age' column:")
print(df['age'].isnull().sum())

# Identify duplicate rows in the dataset
duplicate_rows = df[df.duplicated()]
print(f"\nNumber of duplicate rows: {len(duplicate_rows)}")
if not duplicate_rows.empty:
    print("Sample duplicate rows:")
    print(duplicate_rows.head())
else:
    print("No duplicate rows found.")

# Generate Data Quality Report
print("\n--- Data Quality Report ---")

# Missing Values
print("\nMissing Values per Column:")
print(df.isnull().sum())

# Duplicate Rows
print(f"\nTotal Duplicate Rows: {df.duplicated().sum()}")

# Descriptive Statistics for Numerical Columns
print("\nDescriptive Statistics for Numerical Columns:")
print(df.describe())

# Value Counts for Key Categorical Columns
categorical_cols = df.select_dtypes(include='object').columns
print("\nCategorical Column Distributions:")
for col in categorical_cols:
    print(f"\nColumn: {col}")
    print(df[col].value_counts())

# Optional: Visualize missing values using heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title("Missing Values Heatmap - Titanic Dataset")
plt.show()
```
