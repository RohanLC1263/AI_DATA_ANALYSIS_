```python
# Import required libraries
import pandas as pd
import seaborn as sns

# ----------------------------
# Part 1: Load a Dataset & Check Missing Values
# ----------------------------

# Task 1: Load Iris Dataset and Check for Missing Values
df_iris = sns.load_dataset('iris')
print("First 5 rows of the Iris dataset:")
print(df_iris.head())

# Check for missing values in the entire dataset
print("\nMissing values in the Iris dataset:")
print(df_iris.isnull().sum())

# ----------------------------
# Part 2: Identify Duplicates & Inconsistencies
# ----------------------------

# Task 2: Simulate a dataset with inconsistent 'Gender' entries
sample_data = {
    'Name': ['Alice', 'Bob', 'Cathy', 'Dan', 'Eve', 'Frank'],
    'Gender': ['Male', 'male', 'M', 'Female', 'female', 'F']
}
df_sample = pd.DataFrame(sample_data)

print("\nOriginal 'Gender' column values:")
print(df_sample['Gender'].value_counts())

# Normalize 'Gender' values
df_sample['Gender_cleaned'] = df_sample['Gender'].str.strip().str.lower()
df_sample['Gender_cleaned'] = df_sample['Gender_cleaned'].replace({
    'm': 'male',
    'f': 'female'
})

print("\nCleaned 'Gender' column values:")
print(df_sample['Gender_cleaned'].value_counts())

# ----------------------------
# Part 3: Generate a Data Quality Report
# ----------------------------

# Task 3: Iris Dataset Summary Report
print("\n--- Iris Dataset Summary Report ---")

# Descriptive statistics for numerical columns
print("\nDescriptive Statistics:")
print(df_iris.describe())

# Distribution of categorical column ('species')
print("\nSpecies Distribution:")
print(df_iris['species'].value_counts())
```
